## 5 A Regression Problem

```python
[X_train, Y_train] = generate_dataset_regression(3000, 20)
[X_val, Y_val] = generate_dataset_regression(100, 20)
[X_test, Y_test] = generate_test_set_regression()

input_shape = int(np.sqrt(X_train.shape[1]))
```

```python
X_train = X_train.reshape((X_train.shape[0],input_shape,input_shape,1))
X_val = X_val.reshape((X_val.shape[0],input_shape,input_shape,1))
X_test = X_test.reshape((X_test.shape[0],input_shape,input_shape,1))
```

```python
def ord_summit(y):
    g = (y[0:2]+y[2:4]+y[4:6])/3
    
    x0 = y[0:2]-g
    x1 = y[2:4]-g
    x2 = y[4:6]-g
    
    angle = [0,0,0]
    angle[0] = np.arctan(x0[1]/x0[0]) + np.pi * int(x0[0] < 0)
    angle[1] = np.arctan(x1[1]/x1[0]) + np.pi * int(x1[0] < 0)
    angle[2] = np.arctan(x2[1]/x2[0]) + np.pi * int(x2[0] < 0)
    
    ordre = np.argsort(angle)
    return np.array([y[2*ordre[0]], y[2*ordre[0]+1],y[2*ordre[1]], y[2*ordre[1]+1],y[2*ordre[2]], y[2*ordre[2]+1]  ])

for i in range(Y_train.shape[0]):  
    Y_train[i] = ord_summit(Y_train[i])
    
for i in range(Y_val.shape[0]):  
    Y_val[i] = ord_summit(Y_val[i])
    
for i in range(Y_test.shape[0]):  
    Y_test[i] = ord_summit(Y_test[i])
```

```python
model0 = Sequential()
model0.add(Conv2D(32, 5, input_shape = (input_shape,input_shape,1)))
model0.add(MaxPooling2D())
model0.add(Conv2D(64, 5, input_shape = (input_shape,input_shape,1)))
model0.add(MaxPooling2D())
model0.add(Conv2D(128, 5, input_shape = (input_shape,input_shape,1)))
model0.add(MaxPooling2D())
model0.add(Conv2D(256, 5, input_shape = (input_shape,input_shape,1)))
model0.add(Flatten())
model0.add(Dense(128))
model0.add(Dense(32))
model0.add(Dense(6))

model0.compile(loss='mse', optimizer="adam", metrics=['mse'])
model0.summary()
```

```python
model0.fit(x=X_train, y=Y_train, batch_size=64, epochs=300, validation_data=[X_val, Y_val])
```

```python
Y_pred = model0.predict(X_test)
```

```python
model0.evaluate(X_test, Y_test)
```
